##Validation des donnees
Description du processus de validation et de nettoyage des donnees

1 - L'ensemble de donnees initial contient 120 lignes et 14 colonnes

2- la colonne mois contient 24 valeurs distinct de Janvier 2023 a Decembre 2024

3 - la colonne inscription formation contient 111 valeur distinct dont la minimum est 50 et le maximum est 695

4 - le minimum da la colonne revenu mentuel -150. est et le maximum est 135114

5 - Seul la colonne satisfaction client contient une valeur null qui est a la ligne 59 au mois de Decembre 2023 marque future studio. Cette valeur sera remplacer par la moyenne de la satisfaction de la marque future studio

```python
#remplacement des valeur manquante de satisfaction client au niveau du future studio par la moyenne de la satisfaction de la marque future studio
dataframe.groupby('marque')['satisfaction_clients'].mean()
```

```python
moyenne_satisfaction = dataframe.groupby('marque')['satisfaction_clients'].mean()

dataframe['satisfaction_clients'] = dataframe.apply(
    lambda row: moyenne_satisfaction[row['marque']] if pd.isna(row['satisfaction_clients']) else row['satisfaction_clients'],
    axis=1
)
```
